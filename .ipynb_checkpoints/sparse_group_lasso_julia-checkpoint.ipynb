```julia
# Monte Carlo experiments: OLS vs. LASSO vs. group LASSO vs. sparse group LASSO
using Random, Distributions, LinearAlgebra, Plots, Pythia
rng = MersenneTwister(2020);

M = 5000    # number of MC replications
T = 100     # sample size
p = 100     # number of regressors
K = 50      # iterations for the proximal gradient descend
G0 = 10     # size of first group
G = 5       # other groups
lambda = 1  # regularization parameter
alpha = 0.1 # LASSO and group LASSO weight in [0,1)

bhat_ols = zeros(p, M)
bhat_lasso = zeros(p, M)
Lhat_lasso = zeros(K-1, M)
bhat_glasso = zeros(p, M)
Lhat_glasso = zeros(K-1, M)
bhat_sglasso = zeros(p, M)
Lhat_sglasso = zeros(K-1, M)

beta = vcat([5, 0, 3, 0, 1, 0, 0, 0, 0, 0], zeros(Int, p-10, 1))

# Compute the constant step size
X = randn(rng, Float64, (T, p))
t = 0.5 / sqrt(eigmax(X'*X/T))

@time begin
for m in 1:M
    u = randn(rng, Float64, (T, 1))
    X = randn(rng, Float64, (T, p))
    y = X * beta + u
    bhat_ols[:, m] = X' * X \ X' * y
    
    bhat_lm, Lhat_lm = lasso(X, y, lambda; t = t, K = K)
    bhat_lasso[:, m] = bhat_lm
    Lhat_lasso[:, m] = Lhat_lm
        
    bhat_glm, Lhat_glm = glasso(X, y, lambda, G0, G; t = t, K = K)
    bhat_glasso[:, m] = bhat_glm
    Lhat_glasso[:, m] = Lhat_glm
        
    bhat_sglm, Lhat_sglm = sglasso(X, y, lambda, alpha, G0, G; t = t, K = K)
    bhat_sglasso[:, m] = bhat_sglm
    Lhat_sglasso[:, m] = Lhat_sglm
end
end

```

```julia
# Print results
function l1_error(bhat, beta)
    round(mean(sum(abs.(bhat .- beta), dims = 1)), digits = 2)
end
    
l1_error(bhat_lasso, beta)    
error = [l1_error(bhat_ols, beta), l1_error(bhat_lasso, beta), l1_error(bhat_glasso, beta), l1_error(bhat_sglasso, beta)]'
print("Mean l1 error of OLS, LASSO, group LASSO, and sg-LASSO:", error)

p1 = plot(Lhat_lasso[:, 1:100],  title = "LASSO")
p2 = plot(Lhat_glasso[:, 1:100],  title = "group LASSO")
p3 = plot(Lhat_sglasso[:, 1:100],  title = "sparse group LASSO")
plot(p1, p2, p3, layout = (3, 1), legend = false)
xlabel!("Iteration")
ylabel!("Objective")
```
